## **12) For each year, compute the total ammount of loans. Each loan that has planned expiration time and disburse time in different years must have its amount distributed proportionally to the number of days in each year. For example, a loan with disburse time December 1st 2016, planned expiration time January 30th 2018, and amount 5000 USD, has...**
* 5000 * 31/(31+365+30) = 363.85 for 2016
* 5000 * 365/(31+365+30) = 4284.04 for 2017
* 5000 * 30/(31+365+30)= 352.11 for 2018 

```python
df12 = df2
df12.head(3)
```

```python
print(df12.isna().sum())
```

```python
df12 = df12.dropna(subset=["planned_expiration_time","disburse_time"])
df12 = df12[["loan_amount","disburse_time","planned_expiration_time"]]
```

```python
df12.head(3)
```

```python
df12["start"] = pd.to_datetime(df12["disburse_time"]).dt.year.astype("Int64")
df12["end"] = pd.to_datetime(df12["planned_expiration_time"]).dt.year.astype("Int64")
df12["delta_years"] = df12["end"]-df12["start"]
df12.head()
```

* "first_year_end_days" sono i giorni che mancano alla fine del primo anno
* "last_year_days" sono i giorni che sono passati dall'inizio dell'ultimo anno

```python
df12["first_year_end_days"] = 365 - pd.to_datetime(df12["disburse_time"]).dt.dayofyear
df12["last_year_days"] = pd.to_datetime(df12["planned_expiration_time"]).dt.dayofyear
df12.head()
```

#### Prendo solo quelli il cui delta degli anni è positivo altrimenti non avrebbe senso.

```python
df12 = df12[df12["delta_years"] >= 0]
```

#### Inizializzo le variabili per ciascun anno per andare a inserire la quota pertinente all'anno i-esimo. Laddove non avessi bisogno di tutti gli anni (perché i prestiti possono anche avere origine da dopo il 2011 ed estinguersi prima del 2018) vado a imputare la quota nulla negli anni non utilizzati.

```python
df12b = pd.concat([df12a,years],axis=1,sort=False)
df12b.loc[:,6:] = 0
df12b
```

#### Vado a definire la funzione che inserisce nella colonna corrispondente il valore da attribuire nell'anno *i-esimo*. In particolare, avremo tre casi possibili:
* se il rimborso avviene nello stesso anno del prestito il problema non si pone perché tutta la quota è pertinente allo stesso anno.
* se il prestito avviene nell'anno *i* e il rimborso nell'anno *j = i + 1*, allora dovrò suddividere la quota totale in due tranche. Il totale andrà ponderato rispettivamente per i giorni che mancano alla fine dell'anno *i* e per quelli che sono trascorsi dall'inizio dell'anno *j*.
* se il rimborso avviene a *n > 1* anni dal prestito, ad esempio 2013-2017, non ci saranno solo le due tranche "di coda" del punto precedente ma dovrò imputare una quota pertinente agli *n - 1* anni **interi** per un totale di 365 giorni.

```python
def fx(df):
    
    if df["delta_years"] == 0: 
        df[str(int(df["start"]))] = df["loan_amount"] 
                                                       
    elif df["delta_years"] == 1:
        df[str(int(df["start"]))] = (df["first_year_end_days"] * df["loan_amount"])/((df["first_year_end_days"]) + df["last_year_days"])
        df[str(int(df["end"]))] = df["loan_amount"] - df[str(int(df["start"]))]
        
    elif df["delta_years"] > 1:
        df[str(int(df["start"]))] = df["first_year_end_days"] * df["loan_amount"] / (df["first_year_end_days"] + df["last_year_days"] + ((df["delta_years"] - 1) * 365))
        df[str(int(df["end"]))] = (df["last_year_days"] * df["loan_amount"])/((df["first_year_end_days"]) + df["last_year_days"] + ((df["delta_years"] - 1) * 365))
        
        for i in range(1, int(df["delta_years"])):
            df[str(int(df["start"] + i))] = (365 * df["loan_amount"])/((df["first_year_end_days"]) + df["last_year_days"] + ((df["delta_years"] - 1) * 365))
    
    return df
```

```python
df12b = df12b.apply(fx, axis = 1)
df12b.head(3)
```
